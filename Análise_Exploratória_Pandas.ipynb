```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
plt.style.use("seaborn-v0_8-darkgrid")  # substitua "darkgrid" pelo estilo específico que você deseja
```

```python
caminho = '/content/AdventureWorks-sales-dataset.csv'
df = pd.read_csv(caminho)
df.head(5)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.rename(columns={
    'Product ID': 'Prod_ID',
    'Product Category': 'Prod_Cat',
    'Product Subcategory': 'Prod_Subcat',
    'Product Name': 'Prod_Name',
    'Product Description': 'Prod_Desc',
    'Product Price': 'Prod_Price',
    'Product Weight': 'Prod_Weight',
    'Product Size': 'Prod_Size',
    'Order ID': 'Order_ID',
    'Customer ID': 'Cust_ID',
    'Order Date': 'Order_Date',
    'Order Status': 'Order_Status',
    'Order Quantity': 'Order_Qty',
    'Order Total': 'Order_Total',
    'Payment Method': 'Pay_Method'
}, inplace=True)
```

```python
pd.set_option('display.max_columns', None)  # Mostrar todas as colunas
pd.set_option('display.width', 1000)        # Ajustar a largura da exibição
pd.set_option('display.colheader_justify', 'center')  # Centralizar os nomes das colunas
pd.set_option('display.float_format', '{:.2f}'.format)  # Formatar floats com 2 casas decimais
```

```python
df.head(5)
```

```python
#qual a Receita total?
df['Order_Total'].sum()
```

```python
#qual o valor total de vendas
totaldevendas = df["vendas"] = df['Prod_Price'].mul(df["Order_Qty"])
print(totaldevendas)
```

```python
df.head(1)
```

```python
# Calcular o total de pedidos para cada data
total_por_data = df.groupby('Order Date')['Order_Total'].sum()
print(total_por_data)
```

```python
# Média de vendas
media_vendas = df["Order_Total"].mean()
print("Média de vendas:", media_vendas)
```

```python
#Agrupar a categoria "Prod_Cat" em relação ao total de vendas "Order_Total"
df.groupby("Prod_Cat")["Order_Total"].sum().sort_values(ascending=False)
```

```python
#Gráfico total de produtos vendidos
df.groupby("Prod_Cat")["Order_Total"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Order_Total")
plt.ylabel("Prod_Cat")
```

```python
print(df.columns)
```

```python
if 'Order_Date' in df.columns:
    df['Order_Date'] = pd.to_datetime(df['Order_Date'])
    print("Coluna 'Order_Date' convertida para tipo de dados de data.")
else:
    print("A coluna 'Order_Date' não está presente no DataFrame.")
```

```python
#Número de pedidos por subcategoria de produto

df.groupby('Prod_Subcat')['Order_ID'].count().plot(kind='bar')
```

```python
#Média total de pedidos ao longo do tempo

df.groupby('Order Date')['Order_Total'].mean().plot(kind='line')
```

```python
#Status dos Pedidos

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Order_Status').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
