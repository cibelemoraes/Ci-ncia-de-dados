```python
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

```python
caminho = '/content/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg =carros['mpg']
```

```python
x = range(1, 9)  # Ajustando para ter 8 elementos
y = [1, 2, 3, 4, 5, 3, 2, 1]

plt.bar(x, y)
plt.xlabel('nosso título do eixo x')
plt.ylabel('nosso título do eixo y')
plt.show()
```

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.xlabel('nome dos carros')
plt.ylabel('milhas p/ galão')
plt.title('consumo de gasolina')
plt.xticks(mpg.index, carros.names, rotation=45)

legenda = mpatches.Patch(label='mpg')
plt.legend(handles=[legenda])
```
